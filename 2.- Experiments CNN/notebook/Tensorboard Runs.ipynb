# Tensorboard Runs

```python
import pandas as pd
import numpy as np
import glob, os
from matplotlib import style
import matplotlib.pyplot as plt
import matplotlib

# Matplotlib Config
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8)
matplotlib.rcParams.update({'font.size': 14})
# style.use('ggplot')
```

```python
folder = '/Users/Carlos/Desktop/Magister/5.- Red Convolusional/SSH/Juntas/'
# subfolder = 'SubClases/10Fold/tanh/tensorboard_runs/*.csv'
subfolder = 'Clases/10Fold/Batch 256/tanh/logs/*.csv'


files = []
for file in glob.glob(folder + subfolder):
    files.append(file)
print(len(files))
```

```python
colors = ['r', 'y','g','lightblue','orange','b','brown', 'c', 'm', 'k']

def graph_run(path, num):
    df = pd.DataFrame.from_csv(path).values
    x = df.T[0]
    y = df.T[1]
    name = 'Run ' + str(num)
    plt.scatter(x, y, s=1, color=colors[num%len(colors)], label=name)

plt.rc('axes', axisbelow=True)
plt.grid(color='gray', linestyle='dashed')

for idx, file in enumerate(files):
    graph_run(file, idx + 1)

size=16
plt.ylabel('Accuracy Score', size=size)
plt.xlabel('Epochs', size=size)
# plt.title('Accuracy of the training phase using 10-Fold Cross Validation', size=size)
# plt.ylim(0.5, 0.9)
plt.ylim(0.75, 1)
plt.legend(fancybox=True, loc='lower right', markerscale=5.)
```
